```python
import yfinance as yf
import matplotlib.pyplot as plt

# Get GameStop stock data (ticker symbol: GME)
gme = yf.Ticker("GME")

# Download historical stock prices (you can change the period)
gme_data = gme.history(period="6mo")  # Options: "1mo", "1y", "max", etc.

# Plot the closing price
plt.figure(figsize=(12, 6))
plt.plot(gme_data.index, gme_data['Close'], label='GME Close Price', color='green')
plt.title("GameStop (GME) Stock Price - Last 6 Months")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="6mo")  # you can change to "1y", "max", etc.

# Plot the closing price
plt.figure(figsize=(12, 6))
plt.plot(tesla_data.index, tesla_data['Close'], label='Close Price', color='red')
plt.title("Tesla (TSLA) Stock Price - Last 6 Months")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
